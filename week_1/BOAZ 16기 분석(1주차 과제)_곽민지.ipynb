### 8. Date Column을 이용해 년, 월, 일에 해당하는 3개의 새로운 Column을 만들어주세요.

```python
data['year'] = data['date'].apply(lambda x: x[:4])
data['month'] = data['date'].apply(lambda x: x[4:6])
data['day'] = data['date'].apply(lambda x: x[6:8])
```

### 9. 이후 Price를 예측하는 모델을 제작하기 위해 추가적인 변수를 만드려고 합니다. Price의 데이터 분포를 확인하고 1개 이상의 EDA를 통해 의견을 제시해주세요.

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.rc('font', family='malgun gothic')
plt.rc('axes', unicode_minus=False)
```

```python
# price 변수의 기술통계 확인
data.price.describe().astype('int64')
```

```python
# price 변수의 분포 확인
x = data.price
plt.title("Price")
plt.hist(x, bins=50, edgecolor='w')
plt.show()
```

```python
# data의 상관관계 확인
corr = data.corr(method='pearson')
corr
```

* sqft_living = 0.702899 / grade = 0.667211 / sqft_above = 0.608577 / sqft_living15: 0.586419

```python
#상관관계가 가장 높았던 sqft_living과 price의 산점도 확인
plt.scatter(data.sqft_living, data.price)
plt.show()
```

```python
# 두 번째로 상관관계가 높았던 grade와 price의 산점도 확인
plt.scatter(data.grade, data.price)
plt.show()
```

```python
# grade와 sqft_living의 산점도 확인
plt.scatter(data.grade, data.sqft_living)
plt.show()
```

```python
# grade별 평균 sqft_living 구하기
grade_living = data.groupby('grade')['sqft_living'].agg([('grade_living', 'mean')])
```

```python
grade_living
```

```python
# grade_living을 새로운 변수로 추가
data_merge = pd.merge(data, grade_living, on = 'grade', how='outer')
```

```python
data_merge
```

```python
corr = data_merge.corr(method='pearson')
corr
```

* grade_living과 price의 상관관계는 0.706849로 높은 상관관계를 가지고 있다.
